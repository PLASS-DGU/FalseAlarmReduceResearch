```python
import json
import pandas  as pd
with open("test_bert.json","r") as f:
  test_bert = json.load(f)

with open("train_bert.json","r") as f:
  train_bert = json.load(f)

with open("val_bert.json","r") as f:
  val_bert = json.load(f)

v_test=[]
for i in test_bert:
  v_test.append({'prob_lines':i['prob_lines'],'class_lines':i['class_lines']})

v_train=[]
for i in train_bert:
  v_train.append({'prob_lines':i['prob_lines'],'class_lines':i['class_lines']})

v_val=[]
for i in val_bert:
  v_val.append({'prob_lines':i['prob_lines'],'class_lines':i['class_lines']})

train=pd.read_csv("./src/example_datasets/train.csv")
train['bert_info'] = v_train

test=pd.read_csv("./src/example_datasets/test.csv")
test['bert_info'] = v_test

val=pd.read_csv("./src/example_datasets/val.csv")
val['bert_info'] = v_val
```

```python
import ast
train.line_error=train.line_error.map(ast.literal_eval)
train.error_type=train.error_type.map(ast.literal_eval)
train['type_set']='train'

test.line_error=test.line_error.map(ast.literal_eval)
test.error_type=test.error_type.map(ast.literal_eval)
test['type_set']='test'

val.line_error=val.line_error.map(ast.literal_eval)
val.error_type=val.error_type.map(ast.literal_eval)
val['type_set']='val'
```

```python
all_pd=pd.concat([train, val, test],0)
all_pd.columns=['file','source_code','line_error','error_type','bert_info','type_set']
```

```python
map_tool={
    value:index for index,value in enumerate(ds.tool.unique())
}
# before_1_gt	
# before_1_prob	
# affter_1_gt	
# affter_1_prob	
# before_2_gt	
# before_2_prob	
# affter_2_gt	
# affter_2_prob	
# direct_2_gt	
# direct_2_prob
def get_bert_feature(file, line, bert_info):
  prob_bert = bert_info['prob_lines']
  class_bert = bert_info['class_lines'] # convert ve 0,1
  class_index = class_bert[line-1] #unuse
  class_bert=[min(i,1) for i in class_bert] #min [0,1]

  # hàm dựa vào class_bert và prob_bert để xuất ra bert của line đưa vào
  def get_bert(line, name):
    if line <=0 or line > len(prob_bert):
      return {
          f'{name}_gt':0,
          f'{name}_prob':1.
      }
    return {
        f'{name}_gt':class_bert[line-1],
        f'{name}_prob':prob_bert[line-1]/100.
    }

  feature_bert={}
  # lấy kết quả của 1-2 line kế trước - sau
  for i in [1,2]:
    value = get_bert(line-i, name=f"before_{i}")
    for k,v in value.items():
      feature_bert[k]=v
  
    value = get_bert(line+i, name=f"affter_{i}")
    for k,v in value.items():
      feature_bert[k]=v
  
  # lấy bert của chính line đó
  value = get_bert(line, name=f"direct_{i}")
  for k,v in value.items():
    feature_bert[k]=v
  # class
  return feature_bert

dataset={}

# chạy qua từng file có kết quả từ tool
for index,row in ds.iterrows():
  file  = row.file # file_name
  line = row.line # quét bởi tool

  gt_line = [int(i) for i in row.line_error] # line error
  gt=0

  # nếu 2 kq bằng nhau (tool quét đúng)
  if int(line) in gt_line:
    gt=1

  encode_tool_report_line = f"{file}_{line}"
  if encode_tool_report_line not in dataset:
    bert_feature=get_bert_feature(file, line, row.bert_info) # dict(name:value)
    dataset[encode_tool_report_line]={
        "file":file,
        "line":line,
        "gt":gt,
        'type_set':row.type_set
        
    }
    # gắn giá trị vừa tính các line lân cận
    for name,value in bert_feature.items():
      dataset[encode_tool_report_line][name]=value

    # gắn cho tất cá các tool = 0
    for i in map_tool:
      dataset[encode_tool_report_line][f'tool_{i}']=0

  # gắn riêng tool detect là 1    
  tool = 'tool_' + row.tool
  dataset[encode_tool_report_line][tool] = 1
 
```

```python
dataset=pd.DataFrame(dataset).T
```
